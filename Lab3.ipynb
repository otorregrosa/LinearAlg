```
import numpy as np
import matplotlib.pyplot as plt
```

##TASK 1



$$A = \begin{bmatrix} 7\\28 \\ 54\end{bmatrix} , B = \begin{bmatrix} 5\\9 \\13\end{bmatrix} \\
\\X = 5, Y = 10\\
V = A(X) \cdot B(Y) $$

```
A = np.array([7,28,54])
X = 5
B = np.array([5,9,13])
Y = 10
C = A * X
D = B * Y
V = C @ D
V
```

```
R = np.arange(-10,10,1)
c = np.arange(-10,10,0.0125)

plt.scatter(c*A[0],c*A[1])
plt.scatter(c*B[0],c*B[1])

plt.xlim(-100,100)
plt.ylim(-100,100)

plt.axhline(y=0, color='c')
plt.axvline(x=0, color='y')

plt.grid()
plt.show()
```

```
R = np.arange(-10,10,1)

plt.scatter(c*C[0],c*C[1])
plt.scatter(c*D[0],c*D[1])

plt.xlim(-1000,1000)
plt.ylim(-1000,1000)

plt.axhline(y=0, color='c')
plt.axvline(x=0, color='y')

plt.grid()
plt.show()
```

```
R = np.arange(-10,10,4)

c1, c2 = np.meshgrid(R,R)
vectR = A + B
spanRx = c1*A[0] + c2*B[0]
spanRy = c1*A[1] + c2*B[1]
plt.scatter(spanRx,spanRy, s=5, alpha=0.95)

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.show()
```
