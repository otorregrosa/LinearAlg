```
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

$$P = \begin{bmatrix}-18000\\ 3000\\12000\\10000\end{bmatrix}, R= \begin{bmatrix}1000\\100\\500\\10 \end{bmatrix} \\X = P \cdot R $$

```
profit = np.array([-18000, 3000, 12000, 10000])
reach = np.array([1000, 100, 500, 10])

A = np.array([1000, -18000])
B = np.array([100,3000])
C = np.array([500, 12000])
D = np.array([10, 10000])
M = np.linalg.norm(reach)
N = np.linalg.norm(profit)
theta = math.atan(N/M + 0.5)

plt.xlim( 0, 2000)
plt.ylim( -18000, 8000)

plt.title('Bebang\'s Month Post Efficiency')
plt.xlabel('FB Post Reach Increment')
plt.ylabel('Profit')
plt.quiver(0,0, A[0], A[1], angles='xy', scale_units='xy',scale=1, color='red')
plt.quiver(A[0], A[1], B[0], B[1], angles='xy', scale_units='xy',scale=1, color='b')
R = A + B
plt.quiver (R[0], R[1], C[0], C[1], angles='xy', scale_units='xy',scale=1, color='G')
Q = R + C
plt.quiver(Q[0], Q[1], D[0], D[1], angles='xy', scale_units='xy',scale=1, color='Y')
print(M)
print(N)
print(theta)
plt.grid()
plt.show()
```
